# Importrting Libraries

```python
%matplotlib inline
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import math
import pylab as pl
from sklearn.model_selection import train_test_split as tts
from sklearn.preprocessing import StandardScaler as SS
from sklearn.preprocessing import LabelEncoder as LE
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras import Sequential
import seaborn as sns
```

# Reading the data

```python
df = pd.read_csv("bank.csv",sep = ';')
df.head()
```

# Renaming the Target

```python
df.rename(columns = {'y':'deposit_subscribtion'}, inplace = True)
```

# Getting information about the data

```python
df.describe()
```

```python
df.info()
```

# Saving column names

```python
cols =list(df.columns.values)
cols
```

# Handling unknown data

```python
for col in cols:
    df[col] = df[col].replace("unknown", df[col].mode()[0])
```

# Preparing the data and fitting the model

```python
x = df.drop(["deposit_subscribtion"],axis=1)
y = df["deposit_subscribtion"]
cols = x.columns
for i in range(x.shape[1]):
        if(not (x.iloc[:,i].dtype.kind in 'iufc')):
            x.iloc[:,i] = LE().fit_transform(x.iloc[:,i])
            x.iloc[:,i] = x.iloc[:,i].astype('int')
        else:
            x.iloc[:,i] = SS().fit_transform(x[[x.columns[i]]]).flatten()

y = LE().fit_transform(y)         
#x = pd.DataFrame(x,columns=cols)
#y = pd.DataFrame(y,columns=["deposit_subscribtion"])
XTrain,XTest,YTrain,YTest = tts(x,y, test_size=0.2)
print(YTest)
XTrain = np.asarray(XTrain).astype(np.float32)
XTest = np.asarray(XTest).astype(np.float32)
YTrain = np.asarray(YTrain).astype(np.float32)
YTest = np.asarray(YTest).astype(np.float32)

nn = Sequential()
nn.add(layers.Dense(128, activation='relu', input_shape=(x.shape[1],)))
nn.add(layers.Dense(256, activation='relu', input_shape=(x.shape[1],)))
nn.add(layers.Dense(256, activation='relu', input_shape=(x.shape[1],)))
nn.add(layers.Dense(1, activation='sigmoid'))

print(nn.summary())

nn.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001)
           ,loss=tf.keras.losses.binary_crossentropy
           , metrics=[
            tf.keras.metrics.BinaryAccuracy(name='accuracy'),
            tf.keras.metrics.Precision(name='precision'),
            tf.keras.metrics.Recall(name='recall')
            ])



nn.fit(XTrain, YTrain, validation_data=(XTest,YTest), epochs=1000)

#nn = Sequential()
#nn.add(layers.Dense(32, activation='sigmoid', input_shape=(x.shape[1],)))
#nn.add(layers.Dense(1, activation='linear'))

#print(nn.summary())

#nn.compile(optimizer='sgd',loss='mean_squared_error')



#nn.fit(XTrain, YTrain, validation_data=(XTest,YTest), epochs=100)

```

# Using the model to predict the testing data

```python
preds = nn.predict(XTest).reshape(-1,).round(0)
#segmoinder = lambda t: (1 / (1 + math.exp(-t)))
#segmoind = np.array([segmoinder(xi) for xi in preds])
result = pd.DataFrame(
    {
        'Actual':YTest,
        'Prediction':preds
    }
)

```

# Printing the results of the model

```python
print(result)

result['Resuidual'] = abs(result['Actual']-result['Prediction'])
print(result.iloc[0:99,0].sum())
print(result)

print(result['Resuidual'].sum()/len(result))
```
